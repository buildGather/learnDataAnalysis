```python
#Plotting 
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline

dataset1 = randn(100)
plt.hist(dataset1, color='green')
```

```python
dataset2 = randn(80)
plt.hist(dataset2, color='blue')
```

```python
plt.hist(dataset1, density=True, color='red', alpha=0.4, bins=20)
plt.hist(dataset2, density=True, alpha=0.4, bins=20)
```
